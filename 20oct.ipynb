```python
# Parent class Person, which has two child classes: 1. Student 2. Teacher

class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname
    self.obj1 = Teacher() # component class object in compostite class Person

  def printname(self):
    print(self.firstname, self.lastname)


#Here I have a Student class which is a child class for a Person Class 

class Student(Person):
  def __init__(self, fname, lname, year, age):
    super().__init__(fname, lname)
    self.graduationyear = year
    self.age = age

  def welcome(self):
    print("Welcome", self.firstname, self.lastname, "to the class of", self.graduationyear)
    print(self.age)

  def update(self):
    self.age = self.age +1 
    print(self.age)
      
# Here I have a Teacher class which is a child class for a Person Class 
# child class inheriting parent class 
class Teacher(Person):# child class constructor
  def __init__(self, fname, lname, status):
      super().__init__(fname, lname)
      self.teacherstatus = status
     
      
  def status_check(self): # child class method 
      if (self.teacherstatus == "Mentor"):
           print("her status is : ", self.teacherstatus)
      else: 
          print("She is not able to apply for a job")
          

     
#I have created objects here 
x = Student("Tika", "Ambrosishvili", 2026, 19)
y = Teacher("Khatuna", "Mamukelashvili", "Mentor")
y1 = Teacher("Mary", "Bartia", "unknown")

# calling child class () method 
x.update()

# calling child class Teacher() method 
y1.status_check()

```

```python
##What is Composition (Has-A Relation)? #
class Component: 

# composite class constructor 
	def __init__(self): 
		print('Component class object created...') 

	# composite class instance method 
	def m1(self): 
		print('Component class m1() method executed...') 


class Composite: 

	# composite class constructor 
	def __init__(self): 

		# creating object of component class 
		self.obj1 = Component() 
		
		print('Composite class object also created...') 

	# composite class instance method 
	def m2(self): 
		
		print('Composite class m2() method executed...') 

		# calling m1() method of component class 
		self.obj1.m1() 


# creating object of composite class 
obj2 = Composite() 

# calling m2() method of composite class 
obj2.m2() 
```
